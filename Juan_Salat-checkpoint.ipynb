```python
from math import *
import matplotlib.pyplot as plt
def u(x,t,n):
    suma=0
    for i in range(1,n+1):
        suma=suma+(1-(-1)**i*3)*(2/(i*pi))*sin(i*pi*x)*exp(-(i*pi)**2*t)
    return suma

n=int(input('n='))
xm=float(input('xm='))
xM=float(input('xM='))
tm=float(input('tm='))
tM=float(input('tM='))
np=int(input('NÃºmero de puntos='))
nt=5
for i in range(nt):
    t=tm+i*(tM-tm)/nt
    lx=[]
    lu=[]
    for i in range(np):
        x=xm+i*(xM-xm)/np
        lx.append(x)
        lu.append(u(x,t,n))
    plt.plot(lx,lu,label='t='+"%6.3f"%t)
plt.legend()
plt.xlabel('x')
plt.ylabel('u')
plt.show()
    

        
```
